
### CelebFaces Attributes (CelebA) Dataset

https://www.kaggle.com/datasets/jessicali9530/celeba-dataset => 캐글CelebFaces 속성(CelebA) 데이터 세트 다운받기

- CelebA 데이터셋은 40개의 속성 있는 200,000개 이상의 유명인 이미지가 포함된 대규모 얼굴 데이터셋입니다. 
 > - 10,177 개의 ID 수 
 > - 202,599 개의 얼굴 이미지
 > - 5개의 랜드마크 위치 , 이미지당 40개의 이진 속성 주석.
- CelebA 데이터셋은 약 200,000개 정도의 얼굴 이미지로 구성된다. 
- 기본적으로 10,000명가량의 사람이 포함되어 있다. 즉, 한 명당 20장 정도의 이미지가 있다고 보면 된다. 
- 각 이미지는 178 x 218 해상도로 존재한다.
- 각 얼굴에 대하여 다음과 같은 레이블이 각각 0 혹은 1의 값으로 붙어있다. 확인해 보면 젊은(young), 남성(male), 대머리(bald) 등의 예시가 붙어 있다.




### 남/녀를 구분하여 각 폴더에 저장하기

1. 컬럼명별 폴더 생성하고 각 폴더 안에 -1, 1 폴더 생성


2. male 폴더에 남(1), 여(-1) 이미지 구분하여 각 폴더에 복사하기







```python
## 폴더 제작
import pandas as pd
import os
df=pd.read_csv('./archive/list_attr_celeba.csv')

## list_attr_celeba.csv의 컬럼명별 폴더 제작(/temp 아래)
## 각 폴더 안에 '-1'과 '1' 폴더 만들기
for i in df.columns[1:]:
    os.makedirs(f'./temp/{i}',exist_ok=True)
    os.makedirs(f'./temp/{i}/-1',exist_ok=True)
    os.makedirs(f'./temp/{i}/1',exist_ok=True)

#파일 복사 모듈 (컬러명 Male별 이미지 구분하여 복사하기 )
import shutil

field='Male'
for img,folder in zip(df['image_id'],df['Male']):
    copy=f'./archive/img_align_celeba/img_align_celeba/{img}'
    paste=f'./temp/{field}/{folder}/{img}'
    shutil.copyfile(copy,paste)
```
